# New variables

```r
# overview
head(hourly_int)
```

```r
# parsing and formatting dates and time data, and creating new columns for date and time
sleep_day$SleepDay <- parse_date_time(sleep_day$SleepDay, "%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
sleep_day$date <- format(sleep_day$SleepDay, format="%m/%d/%y")
hourly_int$ActivityHour <- parse_date_time(hourly_int$ActivityHour, "%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
hourly_int$time <- format(hourly_int$ActivityHour, format="%H:%M:%S")
hourly_int$date <- format(hourly_int$ActivityHour, format="%m/%d/%y")
```

```r
# date time separation
hourly_int$ActivityHour = as.POSIXct(hourly_int$ActivityHour, format="%m/%d/%Y %H:%M:%S", tz=Sys.timezone())
hourly_int$time <- format(hourly_int$ActivityHour, format="%H:%M:%S")
hourly_int$date <- format(hourly_int$ActivityHour, format="%m/%d/%y")
hourly_cal$ActivityHour = as.POSIXct(hourly_cal$ActivityHour, format="%m/%d/%Y %H:%M:%S", tz=Sys.timezone())
hourly_cal$time <- format(hourly_cal$ActivityHour, format="%H:%M:%S")
hourly_cal$date <- format(hourly_cal$ActivityHour, format="%m/%d/%y")
daily_activity$ActivityDate = as.POSIXct(daily_activity$ActivityDate, format="%m/%d/%Y", tz=Sys.timezone())
daily_activity$date <- format(daily_activity$ActivityDate, format="%m/%d/%y")
sleep_day$SleepDay = as.POSIXct(sleep_day$SleepDay, format="%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
sleep_day$date <- format(sleep_day$SleepDay, format="%m/%d/%y")
```

# Average total physical intensity vs. Daytime

```r
# creating a new DataFrame to ease the visualization
new_hourly_int <- hourly_int %>%
  group_by(time)%>%
  summarise(mean_total_int = mean(TotalIntensity))
```

```r
# visualization
ggplot(data = new_hourly_int, aes(x=time, y=mean_total_int)) + 
  geom_histogram(stat="identity", fill='darkblue') +
  theme(axis.text.x=element_text(angle=90)) +
  labs(title="Average Total Intensity vs. Time")
```
